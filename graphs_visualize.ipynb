```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
import networkx as nx
```

# COLORS

```python
with open('data/random_graphs_colors_dim3_train.pkl', 'rb') as f:
    train_data = pickle.load(f)
print(train_data.keys(), len(train_data['graph_labels']))
with open('data/random_graphs_colors_dim3_test.pkl', 'rb') as f:
    test_data = pickle.load(f)
print(test_data.keys(), len(test_data['graph_labels']))
```

## COLORS-Train

```python
shapes = np.array([A.shape[0] for A in train_data['Adj_matrices']])
print('COLORS-Train: {} graphs, N_nodes_min={}, N_nodes_max={}'.format(len(shapes), shapes.min(), shapes.max()))
idx = np.where((shapes == 10) & (train_data['graph_labels'] == 4))[0]
for i, ind in enumerate(idx[1:2]):
    G = nx.from_numpy_array(train_data['Adj_matrices'][ind])
    fig = plt.figure()
    nx.draw_networkx(G, node_color=train_data['node_features'][ind], 
                     with_labels=False, width=0.5, edgecolors=[0.3, 0.3, 0.3, 0.1])
    plt.axis('off')    
    plt.show()
    print('N_nodes %d' % train_data['Adj_matrices'][ind].shape[0])
    for key in train_data:
        if isinstance(train_data[key][ind], np.ndarray):
            #print(key, '\n', train_data[key][ind].astype(np.int32))
            pass
        else:
            print(key, train_data[key][ind])
```

## COLORS-Test-Orig

```python
shapes = np.array([A.shape[0] for A in test_data['Adj_matrices']])[:2500]
print('COLORS-Test-Orig: {} graphs, N_nodes_min={}, N_nodes_max={}'.format(len(shapes), shapes.min(), shapes.max()))
idx = np.where((shapes == 10) & (test_data['graph_labels'][:2500] == 4))[0]
for i, ind in enumerate(idx[1:2]):
    G = nx.from_numpy_array(test_data['Adj_matrices'][ind])
    fig = plt.figure()
    nx.draw_networkx(G, node_color=test_data['node_features'][ind], 
                     with_labels=False, width=0.5, edgecolors=[0.3, 0.3, 0.3, 0.1])
    plt.axis('off')    
    plt.show()
    print('N_nodes %d' % test_data['Adj_matrices'][ind].shape[0])
    for key in test_data:
        if isinstance(test_data[key][ind], np.ndarray):
            #print(key, '\n', test_data[key][ind].astype(np.int32))
            pass
        else:
            print(key, test_data[key][ind])
```

## COLORS-Test-Large

```python
shapes = np.array([A.shape[0] for A in test_data['Adj_matrices']])[2500:5000]
print('COLORS-Test-Large: {} graphs, N_nodes_min={}, N_nodes_max={}'.format(len(shapes), shapes.min(), shapes.max()))
idx = np.where((shapes < 35) & (test_data['graph_labels'][2500:5000] == 4))[0] + 2500
for i, ind in enumerate(idx[1:2]):
    G = nx.from_numpy_array(test_data['Adj_matrices'][ind])
    fig = plt.figure()
    nx.draw_networkx(G, node_color=test_data['node_features'][ind], 
                     with_labels=False, width=0.5, edgecolors=[0.3, 0.3, 0.3, 0.1])
    plt.axis('off')    
    plt.show()
    print('N_nodes %d' % test_data['Adj_matrices'][ind].shape[0])
    for key in test_data:
        if isinstance(test_data[key][ind], np.ndarray):
            #print(key, '\n', test_data[key][ind].astype(np.int32))
            pass
        else:
            print(key, test_data[key][ind])
```

## COLORS-Test-LargeC

```python
shapes = np.array([A.shape[0] for A in test_data['Adj_matrices']])[5000:]
print('COLORS-Test-LargeC: {} graphs, N_nodes_min={}, N_nodes_max={}'.format(len(shapes), shapes.min(), shapes.max()))
idx = np.where((shapes < 35) & (test_data['graph_labels'][5000:] == 4))[0] + 5000
for i, ind in enumerate(idx[1:2]):
    G = nx.from_numpy_array(test_data['Adj_matrices'][ind])
    fig = plt.figure()
    node_features = test_data['node_features'][ind].astype(np.float32)
    alpha = node_features.copy()[:, 3]
    alpha[alpha == 1] = 0.5
    alpha[alpha == 0] = 1
    pos = nx.spring_layout(G)
    nx.draw_networkx_edges(G, pos, width=0.5)
    nx.draw_networkx_nodes(G, pos, node_color=node_features, alpha=alpha, 
                           edgecolors=[0.3, 0.3, 0.3, 0.1])    
    plt.axis('off')    
    plt.show()
    print('N_nodes %d' % test_data['Adj_matrices'][ind].shape[0])
    for key in test_data:
        if isinstance(test_data[key][ind], np.ndarray):
            #print(key, '\n', test_data[key][ind].astype(np.int32))
            pass
        else:
            print(key, test_data[key][ind])
```
